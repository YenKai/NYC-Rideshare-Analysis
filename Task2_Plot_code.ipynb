```python
trip_counts = {
        'Uber-1': 13579077,
        'Lyft-1': 6887,
        'Uber-2': 13280761,
        'Lyft-2': 6491,
        'Uber-3': 14554308,
        'Lyft-3': 8444,
        'Uber-4': 13995860,
        'Lyft-4': 8173,
        'Uber-5': 14276372,
        'Lyft-5': 9491
}

         

# Extracting labels and counts from the dictionary
labels = list(trip_counts.keys())
counts = list(trip_counts.values())

# Separate data for Uber and Lyft
uber_labels = [label for label in labels if 'Uber' in label]
uber_counts = [trip_counts[label] for label in uber_labels]

lyft_labels = [label for label in labels if 'Lyft' in label]
lyft_counts = [trip_counts[label] for label in lyft_labels]

# Plotting both Uber and Lyft data on the same histogram with different colors
plt.figure(figsize=(10, 6))

plt.bar(uber_labels, uber_counts, color='skyblue', label='Uber')
plt.bar(lyft_labels, lyft_counts, color='lightgreen', label='Lyft')

plt.xlabel('Business-Month')
plt.ylabel('Trip Counts')
plt.title('Trip Counts by Business-Month')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.yscale('log')  # Use logarithmic scale for y-axis
plt.legend()  # Show legend indicating the colors
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

### Verion 3

```python
import matplotlib.pyplot as plt

# Sample data
trip_counts = {
    'Uber-1': 13579077,
    'Lyft-1': 6887,
    'Uber-2': 13280761,
    'Lyft-2': 6491,
    'Uber-3': 14554308,
    'Lyft-3': 8444,
    'Uber-4': 13995860,
    'Lyft-4': 8173,
    'Uber-5': 14276372,
    'Lyft-5': 9491
}

# Extracting labels and counts for Uber and Lyft
uber_labels = [label for label in trip_counts.keys() if label.startswith('Uber')]
uber_counts = [trip_counts[label] for label in uber_labels]

lyft_labels = [label for label in trip_counts.keys() if label.startswith('Lyft')]
lyft_counts = [trip_counts[label] for label in lyft_labels]

# Plotting both Uber and Lyft data on the same histogram with different colors
fig, ax1 = plt.subplots(figsize=(9, 5))

# Plot Uber data
ax1.bar(uber_labels, uber_counts, color='skyblue', label='Uber')
ax1.set_ylabel('Uber Trip Counts', color='skyblue')

# Create a secondary y-axis for Lyft data
ax2 = ax1.twinx()
ax2.bar(lyft_labels, lyft_counts, color='lightgreen', label='Lyft')
ax2.set_ylabel('Lyft Trip Counts', color='lightgreen')

plt.title('Trip Counts by Business-Month')
plt.xlabel('Business-Month')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Show legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
# Add grid to both axes


plt.show()
```
